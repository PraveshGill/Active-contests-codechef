```python
url="https://www.codechef.com/contests"
```

```python
import requests
```

```python
r=requests.get(url)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(r.content,"html5lib")
```

```python
print(soup.prettify())
```

```python
tables=soup.findAll("table")
```

```python

print(len(tables))
```

```python
table=tables[0]
print(table)
```

```python
tbody=table.findAll('tbody')[1]
#because tbody[1] was useful
```

```python
rows=tbody.findAll('tr')
```

```python
print(len(rows))
```

```python
rows[0]
```

```python
contests=[]
```

```python
for row in rows:
    tds=row.findAll('td')
    contest=[]
    for td in tds:
        contest.append(td.text)
    contests.append(contest)
```

```python
contests
```

```python
from tabulate import tabulate
```

```python
print(tabulate(contests,tablefmt="grid",headers=["code","name","start date","end date"]))
```
